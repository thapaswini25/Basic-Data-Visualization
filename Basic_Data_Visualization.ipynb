```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files

uploaded = files.upload()

df = pd.read_csv(next(iter(uploaded)))

df.head()
```

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
species_mean = df.groupby('Species')['SepalLengthCm'].mean()

plt.figure(figsize=(6,4))

species_mean.plot(kind='bar')

plt.xlabel("Species")

plt.ylabel("Average Sepal Length")

plt.title("Average Sepal Length by Species")

plt.xticks([0,1,2], ['Setosa','Versicolor','Virginica'], rotation=0)

plt.legend(['Sepal Length'])

plt.tight_layout()

plt.savefig("bar_plot.png")

plt.show()
```

```python
plt.figure(figsize=(8,4))

plt.plot(df.index[:30],
         df['SepalLengthCm'][:30],
         marker='o',
         label='Sepal Length')

plt.xlabel("Observation")

plt.ylabel("Sepal Length")

plt.title("Sepal Length Trend")

plt.legend()

plt.tight_layout()

plt.savefig("line_chart.png")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(data=df,
                x='SepalLengthCm',
                y='PetalLengthCm',
                hue='Species')

plt.xlabel("Sepal Length")

plt.ylabel("Petal Length")

plt.title("Sepal Length vs Petal Length")

plt.legend(title='Species',
           labels=['Setosa','Versicolor','Virginica'])

plt.tight_layout()

plt.savefig("scatter_plot.png")

plt.show()
```
